# Question 1) What percentagae of all licensed sets ever released ware Star Wars themed?

```python
merged_df=df.merge(theme,left_on='parent_theme',right_on='name')
merged_df.head()
```

```python
merged_df.drop(columns='name_y',inplace=True)
merged_df.head()
```

```python
licensed=merged_df[merged_df['is_licensed']]
licensed.head()
```

```python
star_wars=licensed[licensed['parent_theme']=='Star Wars']
star_wars.head()
```

```python
merged_df[merged_df['set_num'].isnull()].shape
```

```python
licensed=licensed.dropna(subset='set_num')
```

```python

percentage_StarWars=int((star_wars.shape[0]/licensed.shape[0])*100)
print(f"The percentage of star wars theme among all licensed theme is {percentage_StarWars} %")
```

# Question 2) In which year was Star Wars not the most popular licensed theme ( In terms of number of sets released of that year)?

```python
licensed_sorted=licensed.sort_values('year')
licensed_sorted
```

```python
licensed_sorted['count']=1
```

```python
licensed_sorted
```

```python
summed_df=licensed_sorted.groupby(['year','parent_theme']).sum().reset_index()
summed_df.head()
```

```python
max_df=summed_df.sort_values('count',ascending=False).drop_duplicates(['year'])
```

```python
max_df.sort_values('year',inplace=True)
max_df
```
